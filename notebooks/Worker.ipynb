```python
import schedule
import time
from datetime import datetime

import subprocess

CRAWLER_BOOKS_PATH = ""

def scraper_worker():

    """
    Назначение скрипта: автоматизация парсинга данных с целевого ресурса.
    Условие запуска воркера: время по гринвичу 00:00 + 14.00 часов

    Библиотеки: shedule (schedule-1.2.2)
    """


    cur_datetime = datetime.utcnow().strftime("%Y-%m-%d %H:%M:%S")

    print(f"Запуск парсера: {cur_datetime}")

    try:
        cur_result = subprocess.run(["python3", CRAWLER_BOOKS_PATH])

        # пишу результат в файл
        with open(f"{cur_datetime}", "a", encoding="UTF-8") as work_file:
            work_file.write(cur_result.stdout)

            if cur_result.stderr:
                work_file.write(f"\nОшибки:\n{cur_result.stderr}")
            work_file.write("\n---------------- ----------------\n")

        print(f"Парсер завершил работу: {cur_datetime}")

    except Exception as err:
        print(f"Ошибка в работе скрипта парсера:\n{err}")

schedule.every().day.at("19:00").do(scraper_worker)
print("Воркер активирован.\nСкрипт ожидает запуска наступления 19:00 по UTC")

while True:
    schedule.run_pending()
    time.sleep(60)
```
